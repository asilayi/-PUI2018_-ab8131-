```python
url = "https://data.cityofnewyork.us/resource/hsry-3x74.json"
df = pd.read_json(url)
df.head(3)
```

```python
air_3c = df[['year', 'mean_scale_score', 'num_level_1']]
air_3c.rename(columns = {'year':'Year', 
               'mean_scale_score':'Mean Scale Score', 
               'num_level_1': 'Num Level 1'}, inplace=True)
```

```python
ax = air_3c.plot(x='Year', y='Mean Scale Score');
ax.set_ylabel("Number");
```

```python
ax = air_3c.plot(x='Year', y='Num Level 1')
ax.set_ylabel("Number");
```
